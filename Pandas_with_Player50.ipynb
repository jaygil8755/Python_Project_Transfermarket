```python
df.head()
```

```python
# 컬럼 이름 바꾸고 저장하기 :'team'을 'club'으로 바꾸기 위해 cheat sheet에서 검색 
df.rename(columns = {'team':'club'}, inplace=True)
```

```python
df.head()
```

#### 데이터 전처리(pre-processing)

```python
# value 값에서 불필요한 문자는 없애고 데이터 타입을 숫자형으로 바꾸기
#: €90.00m -> 90.00
# 어떻게? 검색해서...예)) 판다스 특수문자 제거
df['value'] = df['value'].str.replace('€','')
df['value'] = df['value'].str.replace('m','').astype('float')
df.info()
```

#### 칼럼 생성과 삭제

```python
df.head()
```

```python
# 컬럼 생성
#시장가치는 단위가 백만유로인데...13을 곱해 한화로 억원으로 만들어줍니다.
df['시장가치(억)']=df['value']*13
df.head()
```

```python
# 컬럼 삭제 df.drop(columns=['칼럼이름'])
df.drop(columns=['value'], inplace=True)#inplace=True로 해야 저장됨
```

```python
df.head()
```
